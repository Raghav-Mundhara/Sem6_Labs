```python
plt.boxplot(df['turnover'])
plt.show()
```

```python
import numpy as np
q1=np.percentile(df['turnover'],50)
q3=np.percentile(df['turnover'],75)
iqr=q3-q1
print(iqr)
```

```python
lower_bound = q1 -(1.5 * iqr)
upper_bound = q3 +(1.5 * iqr)
data_no_outliers = [x for x in df['turnover'] if (x >= lower_bound) and (x <= upper_bound)]
data_no_outliers
```

```python
plt.boxplot(data_no_outliers)
plt.show()
```
