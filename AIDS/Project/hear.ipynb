```python
new_data=data.drop(['RestingECG','MaxHR','Sex'], axis=1)
```

```python
new_data.head()
```

```python
X=new_data.drop('HeartDisease', axis=1)
y=new_data['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
loggisticModel=LogisticRegression()
loggisticModel.fit(X_train, y_train)
loggisticModel.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFCmodel=RandomForestClassifier()
RFCmodel.fit(X_train, y_train)
RFCmodel.score(X_test, y_test)

```

```python
from sklearn.tree import DecisionTreeClassifier
DTmodel=DecisionTreeClassifier()
DTmodel.fit(X_train, y_train)
DTmodel.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
SVMmodel=SVC()
SVMmodel.fit(X_train, y_train)
SVMmodel.score(X_test, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
NBmodel=GaussianNB()
NBmodel.fit(X_train, y_train)
NBmodel.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNNmodel=KNeighborsClassifier()
KNNmodel.fit(X_train, y_train)
KNNmodel.score(X_test, y_test)
```

```python
all_models=[RFCmodel, loggisticModel, DTmodel, SVMmodel, NBmodel, KNNmodel]
accuracy=[]
for model in all_models:
    accuracy.append(model.score(X_test, y_test))

for i in range(len(all_models)):
    print("%s accuracy is %s"%(all_models[i], accuracy[i]))
```
